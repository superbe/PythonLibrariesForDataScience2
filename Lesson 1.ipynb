#### Задание 1

Привести по 2 примера, когда лучше максимизировать Precision, а когда Recall?

__Точность (Precision)__  
__Precision__ - точность отражает какой процент объектов, отнесённых к "основному" классу, действительно являются таковыми:
1. __Биометрия.__ Здесь важна точность, так как нельзя допустить ошибку при идентификации человека.
2. __Текстовый поиск.__ Для текстового поиска по набору документов точность — это число правильных результатов, деленное на количество всех возвращенных результатов, предполагаемая вероятность того, что случайно выбранный извлеченный документ является релевантным.

__Полнота (Recall)__  
__Recall__ - полнота отражает какой процент объектов "основного" класса было обнаружено алгоритмом:
1. __Тестирование на COVID-19.__ Здесь важна полнота, так как нам важно как можно полнее выявить заболевших.
2. __Проверка безопасности__ (например, обнаружение взрывчатых веществ), здесь так же важна полнота.

#### Задание 2

Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

Мы используем среднее, но не арифметическое, а гармоническое. Рассмотрим на примере (Sasaki, Y. (2007). "The truth of the F-measure").

Среднее арифметическое A (среднее в обычном смысле) и среднее гармоническое H определяются следующим образом:

$$A =\frac{1}{n}\sum_{k=1}^n x_i = \frac{1}{n}(x_1 + x_2 + x_3 + \dots + x_n)$$
$$H =\frac{n}{\sum_{k=1}^n \frac{1}{x_i}} = \frac{n}{(\frac{1}{x_1} + \frac{1}{x_2} + \frac{1}{x_3} + \dots + \frac{1}{x_n})}$$ 


Пусть $x1 = Precision$ и $x2 = Recall$, тогда $A$ и $H$ будут:

$$A = \frac{1}{2}(𝑃𝑟𝑒𝑐𝑖𝑠𝑖𝑜𝑛 + 𝑅𝑒𝑐𝑎𝑙𝑙)$$
$$H = \frac{2}{(\frac{1}{𝑃𝑟𝑒𝑐𝑖𝑠𝑖𝑜𝑛} + \frac{1}{𝑅𝑒𝑐𝑎𝑙𝑙})} = \frac{2}{\frac{𝑃𝑟𝑒𝑐𝑖𝑠𝑖𝑜𝑛 + 𝑅𝑒𝑐𝑎𝑙𝑙}{𝑃𝑟𝑒𝑐𝑖𝑠𝑖𝑜𝑛 \cdot 𝑅𝑒𝑐𝑎𝑙𝑙}} = \frac{2 \cdot 𝑃𝑟𝑒𝑐𝑖𝑠𝑖𝑜𝑛 \cdot 𝑅𝑒𝑐𝑎𝑙𝑙}{𝑃𝑟𝑒𝑐𝑖𝑠𝑖𝑜𝑛 + 𝑅𝑒𝑐𝑎𝑙𝑙} $$


Среднее гармоническое является более интуитивным, чем среднее арифметическое. Предположим, что у нас есть система распознавания отпечатков пальцев и ее точность и полнота будет 1,0 и 0,2 соответственно. Интуитивно понятно, что общая производительность система очень низкой, так как покрывает только 20% зарегистрированных отпечатков пальцев. При этом среднее арифметическое точности и полноты составляет 0,6, а среднее гармоническое

$$\frac{2 \cdot 1 \cdot \frac{2}{10}}{1 + \frac{2}{10}} = \frac{4}{12} = \frac{1}{3}$$

Видно, что среднее гармоническое (0,333) является более разумной оценкой, чем среднее арифметическое (0,6).
