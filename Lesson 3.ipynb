#### Задание 1

Для чего и в каких случаях полезны различные варианты усреднения для метрик качества классификации: micro, macro, weighted?

"micro" подсчитывает общее количество истинных положительных, ложных отрицательных и ложных положительных результатов. Микроусреднение используется в множественной классификации с несколькими метками, где класс большинства меток должен игнорироваться. Используем для многоклассовой классификации.

"macro" рассчитывает метрики для каждой метки и находит их невзвешенное среднее, придавая равный вес каждому классу. В задачах, где есть малочисленные классы (дисбаланс значений классов) и которые тем не менее важны, усреднение может быть средством оценки производительности. С другой стороны, предположение о том, что все классы одинаково важны, часто неверно, так что усреднение будет переоценивать низкую производительность в малочисленном классе. Используем при расчете классификатора со сбалансированными классами.

"weighted" учитывает дисбаланс классов путем вычисления средневзвешенное значения двоичных метрик, в которых оценка каждого класса взвешивается по его присутствию в выборке данных. Используем для классификаторов с явным дисбалансом классов.

#### Задание 2

В чём разница между моделями xgboost, lightgbm и catboost или какие их основные особенности?

Разница в основном в методе сортировки признаков при разбиении на классы при построении модели и способности работать с категориальными признаками.

__Xgboost__

Самая распространенная реализация градиентного бустинга. Создана в 2014 г. Для выбора разбиения используют сортировку и модели, основанные на анализе гистограмм.

__LightGBM__

Вариант градиентного бустинга от Microsoft, вышедший в 2017 г. Для выбора критерия разбиения используется Gradient-based One-Side Sampling (GOSS). Имеются методы работы с категориальными признаками, т.е. с признаками, которые явно не выражаются числом (например, имя автора или марка машины).

__CatBoost__

Разработка компании Яндекс, вышедшаяв 2017 г. Реализует особый подход к обработке категориальных признаков (основанный на target encoding, т.е. на подмене категориальных признаков статистиками на основе предсказываемого значения). К тому же алгоритм содержит особый подход к построению дерева, который показывает лучшие результаты.

На сегодня алгорим CatBoost при соответствующей настройке самый быстрый и точный.
